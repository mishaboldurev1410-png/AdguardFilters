```python
import importlib
import urllib3
import requests
import pandas as pd
import time

APP_ID  = 570060128
COUNTRY = "ru"

print(f"🔑 App ID  : {APP_ID}")
print(f"🌍 Страна  : {COUNTRY}\n")

all_reviews = []

for page in range(1, 11):
    url = (
        f"https://itunes.apple.com/{COUNTRY}/rss/customerreviews/"
        f"page={page}/id={APP_ID}/sortby=mostrecent/json"
    )

    try:
        resp = requests.get(url, timeout=15)
    except Exception as e:
        print(f"Ошибка на странице {page}: {e}")
        break

    if resp.status_code != 200:
        print(f"Страница {page}: статус {resp.status_code}, останавливаемся.")
        break

    data = resp.json()
    entries = data.get("feed", {}).get("entry", [])

    if page == 1 and entries:
        entries = entries[1:]  # первый entry — метаданные приложения

    if not entries:
        print(f"Страница {page}: отзывов больше нет.")
        break

    for e in entries:
        try:
            all_reviews.append({
                "date":   e["updated"]["label"][:10],
                "rating": int(e["im:rating"]["label"]),
                "review": e["content"]["label"]
            })
        except (KeyError, ValueError):
            continue

    print(f"Страница {page}: +{len(entries)} отзывов (всего: {len(all_reviews)})")
    time.sleep(0.5)

# ── Сохранение ──
if not all_reviews:
    print("⚠️  Отзывы не найдены.")
else:
    df = pd.DataFrame(all_reviews)[["date", "rating", "review"]]
    df.to_csv("reviews.csv", index=False, encoding="utf-8-sig")
    print(f"\n✅ Собрано отзывов : {len(df)}")
    print(f"💾 Файл сохранён  : reviews.csv")
```
